Problema 2. Organizar los horarios de partidos de La Liga

Desde la La Liga de fútbol profesional se pretende organizar los horarios de los partidos de liga de cada jornada. Se conocen algunos datos que nos deben llevar a diseñar un algoritmo que realice la asignación de los partidos a los horarios de forma que maximice la

#### ¿Como represento el espacio de soluciones?
Una primera aproximación para representar el espacio de soluciones sería almacenar en listas los equipos que pertenecen a cada categoría. El problema viene a la hora de almacenar los datos que hacen referencia a las ponderaciones por horario y coincidencias, ya que podríamos necesitar muchas variables y sería tedioso trabajar con ellas. Si nos paramos un poco a pensar, nos damos cuenta que la representación óptima de los datos de entrada se estructura en tres diccionarios:
- Categorías: donde la "key" será la categoría y el "value" el nombre del equipo.
- Audiencia base: donde la "key" será la combinación de equipos y el"value" el valor base de la audiencia.
- Ponderación por horario: donde la "key" será el horario y el "value" la ponderación.
- Ponderación por coincidencia: donde la "key" será el número de partidos coincidentes y el "value" la ponderación .

A partir de aquí, deducimos que la solución será un vector de partidos, donde cada partido es a su vez un vector con las keys de los equipos que juegan el partido y el horario.

```
categorias = {'A' : ['Real Madrid', 'Real Sociedad', 'Barcelona'],
           'B' : ['Celta', 'Valencia', 'Athletic', 'Villareal', 'Alavés', 'Levante', 'Espanyol', 'Sevilla', 'Betis', 'Atlético', 'Getafe'],
           'C' : ['Mallorca', 'Eibar', 'Leganés', 'Osasuna', 'Granada', 'Valladolid']}
audiencia_base = {'A-A' : 2, 'A-B' : 1.3, 'A-C' : 1, 'B-A' : 1.3, 'B-B' : 0.9, 'B-C' : 0.75, 'C-A' : 1, 'C-B' : 0.75, 'C-C' : 0.47}
ponderacion_horario = {'V20' : 0.4, 'S12' : 0.55, 'S16' : 0.7, 'S18' : 0.8, 'S20' : 1, 'D12' : 0.45, 'D16' : 0.75, 'D18' : 0.85, 'D20' : 1, 'L20' : 0.4}
ponderacion_coincidencias = {0 : 0, 1 : 0.25, 2 : 0.45, 3 : 0.6, 4 : 0.7, 5 : 0.75, 6 : 0.78, 7 : 0.8, 8 : 0.8}

partidos_jornada = [['Celta', 'Real Madrid'], ['Valencia', 'Real Sociedad'], ['Mallorca', 'Eibar'], ['Athletic', 'Barcelona'], ['Leganés', 'Osasuna'], ['Villareal', 'Granada'], ['Alavés', 'Levante'], ['Espanyol', 'Sevilla'], ['Betis', 'Valladolid'], ['Atlético', 'Getafe']]
```

#### ¿Cual es la función objetivo?

El objetivo del problema es maximizar la audiencia de la jornada, por tanto la función objetivo es el sumatorio de la audiencia base de cada partido multiplicada por la ponderación de horario y la ponderación por coincidencia.

#### ¿Como implemento las restricciones?

Para las restricciones, implementaremos una función que verifique la validez de una solución. Para este problema, las únicas restricciones que debemos satisfacer, y que por tanto serán las que se comprueben en la función de verificación, es la presencia obligatoria de un partido el viernes y un partido el lunes y la no repetición de un mismo partido en la jornada.

#Análisis
- ¿Que complejidad tiene el problema?. Orden de complejidad y Contabilizar el espacio de soluciones

En primer lugar, dado que el problema nos exige la existencia de un partido el lunes y un partido el viernes, calculamos primero las posibles combinaciones entre estos dos días. Para el viernes tenemos 10 posibilidades y, tras elegir una de ellas, nos quedan 9 posibilidades para el lunes, por lo que las combinaciones para estos dos días son 10*9 = 90 combinaciones.

Tras satisfacer las restricciones, nos quedan 8 partidos a repartir entre 8 posibles horarios, pudiendo haber más de un partido por franja horaria. Esto se corresponde a una variación con repetición, y aplicando la fórmula tenemos:

VR = 8<sup>8</sup> = 16.777.216

Finalmente, multiplicamos las combinaciones obtenidas obteniendo un total de 16.777.216*90 = 1.509.949.440 combinaciones.

Observando los cálculos realizados vemos que la complejidad del problema es O(n<sup>m</sup>), donde n es el número de franjas horarias y m el número de partidos.

#Diseño
- ¿Que técnica utilizo? ¿Por qué?

Dado que el problema presenta una complejidad exponencia con un espacio de soluciones muy elevado, debemos optar por un algoritmo que nos permitan tomar atajos o reducir al máximo el número de iteraciones realizadas.

Por ejemplo, podría hacerse uso de un algoritmo voraz que, por etapas, tome las mejores decisiones para la asignación de partidos. No obstante, el uso de algoritmos voraces queda descartado ya que no podemos asegurar que las mejores decisiones parciales nos lleven a la mejor solución final.

Por otro lado, los algoritmos heurísticos permiten el uso de aproximaciones para alcanzar una solución en un tiempo más razonable y, aunque no nos aseguran que esta solución sea la óptima, si proporcionan buenas soluciones cercanas a la óptima. Dentro de los algoritmos heurísticos encontramos, por ejemplo, ramificación y poda que hace uso de aproximaciones para eliminar caminos que no son prometedores, algoritmos de búsqueda basados en trayectorias como el recocido simulado, métodos constructivos como colonias de hormigas, o algoritmos basados en la naturaleza como genéticos y evolutivos.

Dentro de toda esta selección, la opción a implementar es la de un algoritmo genético, ya que es capaz de explorar diferentes regiones del espacio de soluciones en busca de las más óptimas, proporcionando un buen equilibrio entre intensificación (a través del cruce de individuos) y diversificación (a través de mutaciones).

En primer lugar vamos a crear algunas funciones auxiliares para el desarrollo del algoritmo principal.

```
import random
import copy
```

```
# Funcion para determinar la audiencia base en funcion del partido
def encontrar_audiencia_base(partido):
  for key, value in categorias.items():
    if partido[0] in value:
      cat_local = key
    if partido[1] in value:
      cat_visitante = key

  categoria_partido = cat_local + '-' + cat_visitante
  audiencia = audiencia_base[categoria_partido]
  return audiencia
```

```
# Funcion para determinar la audiencia ponderada en base al horario y la coincidencia con otros partidos
def ponderar_partido(partido, horario, coincidencia):
  audiencia = encontrar_audiencia_base(partido)
  audiencia_ponderada = audiencia * ponderacion_horario[horario]
  audiencia_ponderada = audiencia_ponderada * (1 - ponderacion_coincidencias[coincidencia])
  return audiencia_ponderada
```

```
# Funcion para determinar la audiencia total de la jornada
def determinar_audiencia_jornada(jornada):
  horarios = [partido[2] for partido in jornada]
  coincidencias = {}

  # Busco el número de coincidencias en un mismo horario horario
  for horario in horarios:
    if horario in coincidencias:
        coincidencias[horario] += 1
    else:
        coincidencias[horario] = 0

  audiencia = 0
  for partido in jornada:
    audiencia += ponderar_partido(partido[:2], partido[2], coincidencias[partido[2]])

  return audiencia
```

```
# Funcion para crear una distribucion de jornada inicial aleatoria
def crear_jornada_aleatoria(partidos_jornada, horarios_jornada):
  jornada = []
  partidos = copy.deepcopy(partidos_jornada)
  horarios = copy.deepcopy(horarios_jornada)

  # Asignamos en primer lugar un partido al viernes para satisfacer la restricción
  partido = random.choice(partidos)
  jornada += [partido + ['V20']]
  partidos.remove(partido)
  horarios.remove('V20')

  # Asignamos un partido al lunes
  partido = random.choice(partidos)
  jornada += [partido + ['L20']]
  partidos.remove(partido)
  horarios.remove('L20')

  # Asignamos de forma aleatoria el resto de partidos entre los distintos horarios
  for _ in range(len(partidos)):
    partido = random.choice(partidos)
    horario = [random.choice(horarios)]
    jornada += [partido + horario]
    partidos.remove(partido)

  return jornada
```

#### Algoritmo genético

Implementamos ahora las funciones relacionadas con el algoritmo genético

```
# Generamos una población de tamaño N
def generar_poblacion(partidos, horarios, N):
  poblacion = []
  for i in range(N):
    poblacion.append(crear_jornada_aleatoria(partidos, horarios))

  return poblacion
```

```
#Evalua la población y devuelve el mejor individuo
def Evaluar_Poblacion(poblacion):
  mayor_audiencia = 0
  mejor_individuo = poblacion[0]
  for individuo in poblacion:
    audiencia = determinar_audiencia_jornada(individuo)
    if (audiencia > mayor_audiencia):
      mayor_audiencia = audiencia
      mejor_individuo = individuo

  return (mejor_individuo, mayor_audiencia)
```

```
# Para satisfacer las restricciones debemos comprobar que hay un partido el viernes y el lunes, y que no hay partidos repetidos
def Factibilizar(jornada_inicial, partidos_jornada):
  jornada = copy.deepcopy(jornada_inicial)
  lista_partidos = [partido[:2] for partido in jornada]
  lista_horarios = [partido[2] for partido in jornada]

  # Si no hay un partido el viernes elegimos un partido aleatorio y le asignamos el viernes
  if not 'V20' in lista_horarios:
    idx = random.randint(0,len(jornada)-1)
    jornada[idx][2] = 'V20'

  # Repetimos el proceso para el lunes
  if not 'L20' in lista_horarios:
    idx = random.randint(0,len(jornada)-1)
    jornada[idx][2] = 'L20'

  # Buscamos si se repite algún partido y lo sustituimos por el partido que falte
  for idx, partido in enumerate(jornada):
    if(partido[:2] in lista_partidos[idx+1:]):
      for p in partidos_jornada:
        if not p in lista_partidos:
          jornada[idx][:2] = p
          lista_partidos[idx] = p
          break
  return jornada
```

```
#Funcion de mutación. Se eligen dos partidos y se intercambia su horario. Se hacen mutaciones mutacion% de las veces
def Mutar(jornada, mutacion):
  porcentaje = random.randint(0,100)
  if(porcentaje < mutacion * 100):
    partido1 = random.randint(0,len(jornada)-1)
    partido2 = random.randint(0,len(jornada)-1)
    jornada[partido1][2], jornada[partido2][2] = jornada[partido2][2], jornada[partido1][2]

  return jornada
```

```
# Funcion para generar hijos a partir de 2 padres:
# Se elige el metodo de 2-puntos de corte
def Descendencia(padres, mutacion, partidos_jornada):
  p1 = random.randint(1, len(padres[0])-4)
  p2 = random.randint(p1, len(padres[0])-2)
  hijo = padres[0][:p1] + padres[1][p1:p2] + padres[0][p2:]
  hijo = Mutar(hijo, mutacion)
  hijo = Factibilizar(hijo, partidos_jornada)

  return hijo
```

```
# Funcion de cruce. Recibe una poblacion y devuelve la población ampliada con los hijos.
def Cruzar(poblacion, mutacion, partidos_jornada):
  l = len(poblacion)

  for i in range(0,l,2):
    # Para cada par de padres se generan entre 1 y 3 hijos de forma aleatoria
    for j in range(random.randint(1,3)):
      hijo = Descendencia([poblacion[i], poblacion[i+1]], mutacion, partidos_jornada)
      poblacion.append(hijo)

  return poblacion
```

```
# Funcion de seleccion de la población. Dada una población, aseguramos que pasan a la siguiente generación los mejores individuos.
# El número de inviduos que avanzan viene dado por la variable elitismo.
# El resto de la población hasta tener N individuos se elige de manera aleatoria, pero los mejores individuos tendrán más oportunidades
def Seleccionar(poblacion_inicial, N, elitismo):
  poblacion = copy.deepcopy(poblacion_inicial)
  num_elite = int(len(poblacion) * elitismo)
  nueva_poblacion = []

  for _ in range(num_elite):
    (mejor_individuo, _) = Evaluar_Poblacion(poblacion)
    nueva_poblacion.append(mejor_individuo)
    poblacion.remove(mejor_individuo)

  pesos = []
  for individuo in poblacion:
    pesos.append(determinar_audiencia_jornada(individuo))

  while(len(nueva_poblacion) < N):
    idx = random.choices(range(len(poblacion)), weights=pesos)[0]
    individuo_aleatorio = poblacion[idx]
    nueva_poblacion.append(individuo_aleatorio)
    poblacion.pop(idx)
    pesos.pop(idx)

  return nueva_poblacion
```

```
def algoritmo_genetico(partidos_jornada, horarios, N=100,mutacion=.15,elitismo=.1,generaciones=100):

  #Genera la poblacion inicial
  poblacion = generar_poblacion(partidos_jornada, horarios, N)

  #Inicializamos valores para la mejor soluci on
  (mejor_solucion, mejor_audiencia) = Evaluar_Poblacion(poblacion)

  n=0
  mejor_solucion_total = mejor_solucion
  mejor_audiencia_total = mejor_audiencia
  cont_mejora = 0

  # Inciamos el cliclo de generaciones
  while(n < generaciones) :

    # Cruce de la poblacion(incluye mutación)
    poblacion = Cruzar(poblacion, mutacion, partidos_jornada)

    # Seleccionamos la población
    poblacion = Seleccionar(poblacion, N, elitismo)

    # Evaluamos la nueva población
    (mejor_solucion, mejor_audiencia) = Evaluar_Poblacion(poblacion)

    print("Generacion #", n, "\nLa mejor solución es:" , mejor_solucion, "\ncon audiencia " , mejor_audiencia, "\n")

    # Numero de generaciones. Criterio de parada
    n +=1

    if mejor_audiencia > mejor_audiencia_total:
      mejor_audiencia_total = copy.deepcopy(mejor_audiencia)
      mejor_solucion_total = copy.deepcopy(mejor_solucion)
      cont_mejora = 0
    else:
      cont_mejora += 1

    # Segunda condicion de parada. Si el algoritmo no mejora en un tercio de las generaciones asumimos que se ha estancado y detenemos la ejecucion
    if cont_mejora == int(generaciones/3):
      break

  return mejor_solucion_total


sol = algoritmo_genetico(partidos_jornada=copy.deepcopy(partidos_jornada), horarios=list(ponderacion_horario.keys()), N=100, mutacion=.3, elitismo=.40, generaciones=250)

print('--------------------------')
print(f"La mejor solución encontrada es {sol} \ncon una audiencia de {determinar_audiencia_jornada(sol)}")
```
